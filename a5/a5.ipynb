```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
life_df = pd.read_csv('/content/Life_Expectancy_Data.csv')
life_df.fillna(0, inplace=True)
life_df.head(3)
```

```python
sns.violinplot(data=life_df, y='Polio', inner='quart').set_title('% Polio inoculation Coverage')
```

```python
sns.violinplot(data=life_df, y='Diphtheria ', inner='quart').set_title('% Diphtheria inoculation Coverage')
```

```python
heat_df = life_df.drop(['Country', 'Year', 'Status'], axis=1)
corr = heat_df.corr()
sns.heatmap(corr, cmap='RdBu')
```
